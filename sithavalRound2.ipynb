```python
import pytesseract
from pdf2image import convert_from_path
import os

# Set Tesseract path (Google Colab should already have it installed, but you can check)
pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"  # This is the Tesseract location in Colab

# Function to extract text from images in the PDF
def extract_text_from_pdf(pdf_path):
    try:
        # Convert PDF to images (high DPI for better quality OCR)
        pages = convert_from_path(pdf_path, 300)  # You can adjust DPI for better quality
        extracted_text = ""

        # Extract text from each page/image using OCR
        for page in pages:
            extracted_text += pytesseract.image_to_string(page)

        if not extracted_text:
            print("No text was extracted from the PDF. Please check the quality of the document.")

        return extracted_text
    except Exception as e:
        print(f"Error in extracting text from PDF: {str(e)}")
        return ""

# Example usage
pdf_file_path ='/content/drive/MyDrive/Tables, Charts, and Graphs with Examples from History, Economics, Education, Psychology, Urban Aff.pdf' # Path to the PDF file in your Colab environment

# Extract text from PDF
extracted_text = extract_text_from_pdf(pdf_file_path)
print(extracted_text[:500])  # Print the first 500 characters for preview
```

```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader

loaders = [PyPDFLoader('/content/drive/MyDrive/Tables, Charts, and Graphs with Examples from History, Economics, Education, Psychology, Urban Aff.pdf')]

docs = []

for file in loaders:
    docs.extend(file.load())

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
docs = text_splitter.split_documents(docs)
embedding_function = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2", model_kwargs={'device': 'cpu'})

vectorstore = Chroma.from_documents(docs, embedding_function, persist_directory="./chroma_db_nccn")

print(vectorstore._collection.count())
```
